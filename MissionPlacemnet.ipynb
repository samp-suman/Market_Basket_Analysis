## Q1 : Reverse the  Array -
https://www.geeksforgeeks.org/write-a-program-to-reverse-an-array-or-string/

```python
array = [1000, 11, 445, 1, 330, 3000]

def reverse(array):
  start=0
  end=len(array)-1
  while start<end:
    array[start], array[end] = array[end], array[start]
    start+=1
    end-=1
  return array
reverse(array)
```

## Q2 : Find the maximum and minimum element in an array -
https://www.geeksforgeeks.org/maximum-and-minimum-in-an-array/

```python
array = [-00, 11, 445, -1, 330, 3000]

def minMax(array):
  mn=array[0]
  mx=array[0]
  for i in array:
    mn = min(i, mn)
    mx = max(i, mx)
  return mn, mx
minMax(array)
```

## Q3 : Find the "Kth" max and min element of an array
https://practice.geeksforgeeks.org/problems/kth-smallest-element/0

```python
array = [-00, 11, 445, -1, -1, 330, 330, 300]

def kth_max(array, k):
  array = sorted(set(array))
  return array[-k]
def kth_min(array, k):
  array = sorted(set(array))
  return array[k-1]

# Duplicates values dicarding
print(kth_max(array, 2),  kth_min(array, 2))
```

## Q4 : Given an array which consists of only 0, 1 and 2. Sort the array without using any sorting algo
https://practice.geeksforgeeks.org/problems/sort-an-array-of-0s-1s-and-2s/0

```python
# Three pointer Aproach

# Array
ar = [0,0,1,2,1,2,0,1,2,0,0]

n = len(ar)

p0 = 0
p1 = 0
p2 = n-1

while p1<p2:
  if ar[p1]==0:
    ar[p1], ar[p0] = ar[p0], ar[p1]
    p0+=1
    p1+=1
  elif ar[p1]==1:
    p1+=1
  else:
    ar[p2], ar[p1] = ar[p1], ar[p2]
    p2-=1
print(ar)
```
